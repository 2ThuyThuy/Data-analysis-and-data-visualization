# SF Salaries Exercise

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Salaries.csv')
```

```python
df.head()
```

```python
# average BasePay
clear = list(df[df['EmployeeName'] == 'Not provided'].index)
clear
```

```python
df.drop(clear,axis=0,inplace=True)
```

```python
df['BasePay']=df['BasePay'].astype(float)
df['BasePay'].mean()
```

```python
#the highest amount of OvertimePay in the dataset
df['OvertimePay']=df['OvertimePay'].astype(float)
df['OvertimePay'].max()
```

```python
#What is the job title of JOSEPH DRISCOLL? 
df[df['EmployeeName'] =='JOSEPH DRISCOLL']['JobTitle']
```

```python
#How much does JOSEPH DRISCOLL make (including benefits)
df[df['EmployeeName'] =='JOSEPH DRISCOLL']['TotalPayBenefits']
```

```python
#What is the name of highest paid person(including benefits)
df[df['TotalPayBenefits'] ==df['TotalPayBenefits'].max()]['EmployeeName']
```

```python
#What is the name of highest paid person(including benefits)
#
df.loc[df['TotalPayBenefits'].idxmax()]
```

```python
#What is them name of lowest paid person (including benefits)? 
#Do you notice something starnage about how much he or she is paid?
df[df['TotalPayBenefits'] ==df['TotalPayBenefits'].min()]['EmployeeName']
```

```python
#What was the average (mean)basePay of all employses per year? (2011-2014)
df.groupby('Year').mean()['BasePay']
```

```python
#How many unique job titles are there
#df['JobTitle'].unique() # array
df['JobTitle'].nunique() 
```

```python
#What are the top 5 most common jobs
df['JobTitle'].value_counts().head(5)
```

```python
#How many Job Titles were represented by only one person in 2013?
#JobTitles with only once occurence in 2013
sum(df[df['Year'] == 2013]['JobTitle'].value_counts() == 1)
```

```python
#How many people have the word Chief in their job tile?(This is pretty tricky)
def chief_string(title):
  if 'chief' in title.lower().split():
    return True
  return False
```

```python
sum(df['JobTitle'].apply(lambda x: chief_string(x)))
```

```python
#Is there a correlation between length of the job Title string and Salary?
df['title_len']=df['JobTitle'].apply(len)
df['title_len']
```

```python
df[['TotalPayBenefits','title_len']].corr() #Compute pairwise correlation of columns
```
